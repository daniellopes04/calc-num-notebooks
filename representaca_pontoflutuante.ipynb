```julia
A(h) = max(abs((sqrt(1 + h) - 1) / h - 0.5), 1e-10)
E(h) = max(abs(1 / (sqrt(1 + h) + 1) - 0.5), 1e-16)
valores_h = [2.0^p for p = range(-50, -10, length=100)]

plot(xaxis=:log, yaxis=:log)
scatter!(valores_h, A.(valores_h), c=:red, ms=3)
scatter!(valores_h, E.(valores_h), c=:blue, ms=3)
```
